## Descripción

Este proyecto desarrolla un **estudio experimental del aprendizaje por refuerzo en entornos secuenciales complejos**, donde las decisiones del agente dependen del estado actual del entorno y las recompensas futuras.

A diferencia del problema del **bandido de k-brazos**, en estos entornos la interacción se modela mediante **Procesos de Decisión de Markov (MDP)**, donde el agente aprende una política óptima a partir de su experiencia sin conocer previamente el modelo del entorno.

Se estudian distintos enfoques de aprendizaje por refuerzo:

### Métodos Tabulares

* **Monte Carlo**

  * On-policy
  * Off-policy
* **Diferencias Temporales**

  * SARSA y expected SARSA
  * Q-Learning y Double Q-Learning

### Métodos con Aproximación de Funciones

* **SARSA semi-gradiente** (con tilings y con RNA)
* **Deep Q-Learning**

Los experimentos se realizan utilizando la librería **Gymnasium**, que proporciona entornos para evaluar estos algoritmos.

El objetivo del proyecto es analizar empíricamente:

* La **evolución de la recompensa acumulada por episodio**.
* La **duración media de los episodios**.
* La **estabilidad del aprendizaje**.
* La **convergencia de las funciones de valor**.
* El comportamiento de los agentes en **espacios de estados discretos y continuos**.
* El impacto de **hiperparámetros** como:
  * tasa de aprendizaje $\alpha$
  * factor de descuento $\gamma$
  * estrategias de exploración $\epsilon$-greedy

Se estudian distintos entornos con características complementarias:

* **FrozenLake**: entorno discreto sencillo que permite analizar algoritmos Monte Carlo.
* **Taxi-v3**: entorno tabular con dinámica más compleja para estudiar métodos de diferencias temporales.
* **LunarLander-v3**: entorno con espacio de estados continuo, que requiere el uso de aproximación de funciones.
